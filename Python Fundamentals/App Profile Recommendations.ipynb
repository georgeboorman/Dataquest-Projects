```python
#We are looking for duplicate apps in the Google Play Store dataset#
duplicate_apps = []
unique_apps = []
for app in android:
    name = app[0]
    if name in unique_apps:
        duplicate_apps.append(name)
    else:
        unique_apps.append(name)

#Below we'll find out how many duplicate apps there are, and display some examples:#
print('Number of duplicate apps:', len(duplicate_apps))
print('\n')
print('Examples of duplicate apps:', duplicate_apps[:6])
```

As we have found a total of 1,181 duplicate apps, our next task is to find a systematic way of cleaning the data by removing multiple entries of apps.

As most categories e.g. genre, price, last updated etc are likely to have the same data across duplicates, a good way to disgtinguish which data to remove is by using the number of reviews. If we write code to look for the highest number of reviews across duplicate apps, and remove all the others, we will only be left with the most popular apps. 

```python
print('Expected number of apps:', len(android) - 1181)
```

```python
reviews_max = {}
for app in android:
    name = app[0]
    n_reviews = float(app[3])
    
    if name in reviews_max and reviews_max[name] < n_reviews:
        reviews_max[name] = n_reviews
    elif name not in reviews_max:
        reviews_max[name] = n_reviews

print('Expected length:', len(android) - 1181)
print('Actual length:', len(reviews_max))
```

```python
android_clean = []
already_added = []
for app in android:
    name = app[0]
    n_reviews = float(app[3])
    if (reviews_max[name] == n_reviews) and (name not in already_added):
        android_clean.append(app)
        already_added.append(name)
```

```python
explore_data(android_clean, 0, 3, True)
```

The steps taken above were:
- Check how many unique apps wehave within the Google Play Store dataset.
- Create an empty dictionary. We then iterated over the data to store apps as keys, against their highest volume of rating (ignoring any incidences of the same app with ratings lower than the largest found. We printed the length of this list against our expected length to confirm they match.
- We created an empty list to store a clean dataset, and another to just store app names.
- We looped through the dataset to check if the number of ratings for an app matched the previously created dictionary and, if so, stored in the list.

```python
def is_english(string):
    non_ascii = 0
    
    for character in string:
        if ord(character) > 127:
            non_ascii += 1
   
    if non_ascii > 3:
        return False
    else:
        return True
    
print(is_english('Docs To Go‚Ñ¢ Free Office Suite'))
print(is_english('Instachat üòú'))
print(is_english('Áà±Â•áËâ∫PPS -„ÄäÊ¨¢‰πêÈ¢Ç2„ÄãÁîµËßÜÂâßÁÉ≠Êí≠'))
```

```python
android_english = []
ios_english = []

for app in android_clean:
    name = app[0]
    if is_english(name):
        android_english.append(app)

for app in ios:
    name = app[2]
    if is_english(name):
        ios_english.append(app)

explore_data(android_english, 0, 3, True)
print('\n')
explore_data(ios_english, 0, 3, True)
```

By reviewing a sample of our datasets, we can identify that the following columns need to be isolated to build a frequency table informing our app development:
- For iOS column 12 `prime_genre`
- For Android columns 1 and 9 `Category` and `Genres` respectively


```python
def freq_table(dataset, index):
    table = {}
    total = 0
    
    for row in dataset:
        total += 1
        value = row[index]
        if value in table:
            table[value] += 1
        else:
            table[value] = 1

    table_percentages = {}
    for key in table:
        percentage = (table[key] / total) * 100
        table_percentages[key] = percentage
    
    return table_percentages

def display_table(dataset, index):
    table = freq_table(dataset, index)
    table_display = []
    for key in table:
        key_val_as_tuple = (table[key], key)
        table_display.append(key_val_as_tuple)

    table_sorted = sorted(table_display, reverse = True)
    for entry in table_sorted:
        print(entry[1], ':', entry[0])

ios_prime_genre = print(display_table(ios_free_apps, -5))
print('\n')
android_genre = print(display_table(android_free_apps, -4))
```

However before we proceed, we need to understand not only the frequency of apps by genre across both markets, but also how many users there are for each category. We will start with the App Store dataset, as this doesn't have a column for the number of installations we will use the total number of user ratings as a proxy. 

```python
genres_ios = freq_table(ios_free_apps, -5)

for genre in genres_ios:
    total = 0
    len_genre = 0
    for app in ios_free_apps:
        genre_app = app[-5]
        if genre_app == genre:
            n_ratings = float(app[6])
            total += n_ratings
            len_genre += 1
    average_n_ratings = total / len_genre
    print(genre, ':', average_n_ratings)
```

From the above we can see the following categories have a high volume of user ratings: `Navigation`, `Reference`, `Social Networking`, `Music` and `Weather`. `Social Networking` is the most popular category of those listed here, but still only accounts for around 3% of iOS apps and will be heavily influenced by Facebook, WhatsApp and others:

```python
for app in ios_free_apps:
    if app[-5] == 'Social Networking':
        print(app[2], ':', app[6])
```

Therefore we should consider the categories, such as `Books` and `Reference` as whilst the markets are smaller the engagement of these types of apps looks quite high. We can consider the creation of an app which spans across these two categories, for example an app which includes a book plus additional features to make it more interactive. 

```python
categories_android = freq_table(android_free_apps, 1)
for category in categories_android:
    total = 0
    len_category = 0
    for app in android_free_apps:
        category_app = app[1]
        if category_app == category:
            installs = app[5]
            n_installs = installs.replace('+', '')
            n_installs = n_installs.replace(',', '')
            n_installs = float(n_installs)
            total += n_installs
            len_category += 1
    avg_installs = total / len_category
    print(category, ':', avg_installs)
```

Based on the Google Play installations, with reference to the App Store analysis conducted previously, we can further conclude that `Books and Reference` would be a suitable profile for a potentially profitable app across both stores.
